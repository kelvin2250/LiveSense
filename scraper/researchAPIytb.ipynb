```python
import os
from googleapiclient.discovery import build
from dotenv import load_dotenv
import json

load_dotenv()
API_KEY = os.getenv("YOUTUBE_API_KEY")

# 1. Khởi tạo client
youtube = build("youtube", "v3", developerKey=API_KEY)

# DEBUG: Làm sao biết 'youtube' object này có thể làm được gì?
print("--- CÁC PHƯƠNG THỨC CỦA YOUTUBE CLIENT ---")
print(dir(youtube)) 
# Bạn sẽ thấy các từ như 'videos', 'liveChatMessages', 'channels'...
```

```python
VIDEO_ID = "ORhUy0xkmYM"

# Gọi API lấy thông tin video
request = youtube.videos().list(
    part="liveStreamingDetails",
    id=VIDEO_ID
)
response = request.execute()
print(response.keys())  # Kiểm tra kiểu dữ liệu của response
# DEBUG: Đừng vội code tiếp, hãy xem 'response' thực sự trông như thế nào
print("--- CẤU TRÚC RESPONSE CỦA VIDEO LIST ---")
print(json.dumps(response, indent=2)) 

# Dựa vào kết quả in ra, bạn sẽ thấy đường dẫn:
# items -> [0] -> liveStreamingDetails -> activeLiveChatId
chat_id = response['items'][0]['liveStreamingDetails']['activeLiveChatId']
print(f"Chat ID tìm được: {chat_id}")
```

```python
chat_request = youtube.liveChatMessages().list(
    liveChatId=chat_id,
    part="snippet,authorDetails"
)
chat_response = chat_request.execute()

# DEBUG: Soi một tin nhắn mẫu (item đầu tiên)
if chat_response['items']:
    sample_item = chat_response['items'][0]
    print("--- CẤU TRÚC MỘT TIN NHẮN ---")
    print(json.dumps(sample_item, indent=2))
    print("--- NỘI DUNG TIN NHẮN ---")
    print(sample_item['snippet']['displayMessage'])
    print("--- TÊN NGƯỜI GỬI ---")
    print(sample_item['authorDetails']['displayName'])
```
